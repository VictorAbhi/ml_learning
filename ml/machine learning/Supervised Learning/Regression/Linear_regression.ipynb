```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

quantity=[1,5,7,9,10,11]
price=[300,400,500,600,800,1000]

dataframe=({
    'quantity':quantity,
    'price':price
})
print(dataframe)
```

```python
# step 1: convert lists to numpy arrays for easier calculations
x = np.array(dataframe['quantity'])
y = np.array(dataframe['price'])

# step 2: Calculate the means of x and y
x_mean = np.mean(x)
y_mean = np.mean(y)

# step 3: calculate the slope(m)
numerator = np.sum((x-x_mean)*(y-y_mean))
denominator = np.sum((x-x_mean)**2)
m = numerator / denominator
b=y_mean - m*x_mean

#print the model: y= mx+b
print(f"Linear Regression model using least squares methd: y= {m:.2f}x+{b:.2f}")

# step 5: Predict y values using the model
y_pred = m*x+b
print("predicted values of y:", y_pred)
print("Total error:",np.sum((y - y_pred)**2))
```

```python
plt.scatter(x, y, color='purple', label='Data Points')
plt.plot(x, y_pred, color='red', label='Linear Regression Line')

plt.xlabel('Quantity')
plt.ylabel('Price')
plt.title('Scatter Plot with Linear Regression Line')
plt.legend()

plt.grid(True)
plt.show()
```
