```python
import csv

class DietRecommendation:
    def __init__(self, age: int, height: float, weight: float, activity_level: str, body_type: str):
        self.age = age
        self.height = height
        self.weight = weight
        self.activity_level = activity_level
        self.body_type = body_type
        self.bmr = self.calculate_bmr()

    def calculate_bmr(self) -> float:
        """Calculate the Basal Metabolic Rate (BMR) based on body type."""
        if self.body_type == 'endomorphic':
            return (10 * self.weight) + (6.25 * self.height) - (5 * self.age) - 161
        elif self.body_type == 'ectomorphic':
            return (10 * self.weight) + (6.25 * self.height) - (5 * self.age) + 5
        elif self.body_type == 'mesomorphic':
            return (10 * self.weight) + (6.25 * self.height) - (5 * self.age) + 100
        else:
            raise ValueError("Invalid body type. Choose from 'endomorphic', 'ectomorphic', or 'mesomorphic'.")

    def calorie_calculate(self, aim: str) -> float:
        """Calculate calorie needs based on activity level and goal."""
        bmr = self.bmr
        activity_multipliers = {
            'sedentary': 1.2,
            'lightly_active': 1.375,
            'moderately_active': 1.55,
            'very_active': 1.725,
            'extremely_active': 1.9
        }

        if self.activity_level not in activity_multipliers:
            raise ValueError("Invalid activity level.")
        bmr *= activity_multipliers[self.activity_level]

        if aim == 'weight_loss':
            return bmr * 0.8
        elif aim == 'weight_gain':
            return bmr * 1.2
        elif aim == 'maintain':
            return bmr
        else:
            raise ValueError("Invalid goal. Choose from 'weight_loss', 'weight_gain', or 'maintain'.")

    def display_calorie_intake_for_goal(self, aim: str):
        """Display the total calorie intake for the specified goal."""
        total_calories = self.calorie_calculate(aim)
        print(f"Total calorie intake for {aim.replace('_', ' ').title()}: {total_calories:.2f} calories/day")

    def load_food_items(self, csv_file: str) -> list:
        """Load food items from a CSV file."""
        food_items = []
        try:
            with open(csv_file, 'r') as file:
                reader = csv.DictReader(file)
                for row in reader:
                    food_items.append(row)
        except FileNotFoundError:
            raise FileNotFoundError(f"File '{csv_file}' not found.")
        return food_items

    def select_meal(self, food_items: list, target_cal: float) -> dict:
        """Select the meal closest to the target calorie count."""
        closest_meal = None
        closest_distance = float('inf')
        for meal in food_items:
            cal_count = int(meal['Calories'])
            distance = abs(cal_count - target_cal)
            if distance < closest_distance:
                closest_meal = meal
                closest_distance = distance
        return closest_meal
```

```python
diet =DietRecommendation(age=25, height=175, weight=70, activity_level='moderately_active', body_type='mesomorphic')
goal = input("Enter your goal(weight_gain, weight_loss, maintain)").strip()
diet.display_calorie_intake_for_goal(goal)

breakfast_items = diet.load_food_items('data_sets/Breakfast_data.csv')
dinner_items = diet.load_food_items('data_sets/Dinner_data.csv')
lunch_items = diet.load_food_items('data_sets/Lunch_data.csv')

target_calories = diet.calorie_calculate(goal)
breakfast_meal = diet.select_meal(breakfast_items,target_calories)
lunch_meal = diet.select_meal(lunch_items,target_calories)
dinner_meal = diet.select_meal(dinner_items,target_calories)

print(f"\n Recommended breakfast : {breakfast_meal['Food_items']} {breakfast_meal['Calories']}.calories")
print(f"\n Recommended lunch : {lunch_meal['Food_items']} {lunch_meal['Calories']}.calories")
print(f"\n Recommended Dinner : {dinner_meal['Food_items']} {dinner_meal['Calories']}.calories")
```
