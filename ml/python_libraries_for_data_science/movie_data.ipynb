```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('data_sets/tmdb_5000_movies.csv')
df2 = pd.read_csv('data_sets/tmdb_5000_credits.csv')
```

```python
df2.rename(columns={'movie_id':'id'}, inplace=True)
```

```python
df = pd.merge(df1, df2, on='id', how='inner')
```

```python
df.to_csv('movies.csv', index=False)
df.isnull().sum()
```

```python
genre_count = {}
for g in df['genres']:
    for genre in g:
        name = genre['name']
        if name in genre_count:
            genre_count[name] += 1
        else:
            genre_count[name] = 1

sorted_genres = sorted(genre_count.items(), key=lambda x: x[1], reverse=True)
print(sorted_genres[:10])
```

```python
# visualize genres count
genres = [i[0]for i in sorted_genres]
count = [i[1]for i in sorted_genres]

plt.figure(figsize=(10,6))
sns.barplot(x=count,y=genres, palette='viridis')

plt.xlabel('Count', fontsize=12)
plt.ylabel('Genre', fontsize=12)
plt.title('Genres by Count', fontsize=15)
plt.tight_layout()

plt.show()
```
