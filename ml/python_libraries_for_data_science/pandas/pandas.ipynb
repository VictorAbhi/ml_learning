```python
def double_make(s):
    return s*2;
```

```python
df[['sepal_length','petal_width']].apply(double_make)
```

```python
df.sort_values(by='petal_width') # sorting in the ascending order
```

```python
df.sort_values(by='petal_width', ascending=False) # sorting in the descending order
```

```python
# sort by multiple columns
sorted_data_multiple = df.sort_values(by=['sepal_length', 'sepal_width'], ascending=[False, True])
print("\nSorted data by sepal_length (descending) and sepal_width (ascending):")
print(sorted_data_multiple.head())
```

```python
# Filtering the dataset
# select rows where petal_lenght > 1.5
filtered_data = df[df['petal_length'] > 2.5]
print( "\nFiltered data (petal_length > 2.5):")
print(filtered_data.head())
```

```python
# Grouping by class and calculating means
grouped_data = df.groupby('class').mean()
print("\nMean values grouped by species:")
print(grouped_data)
```

```python
# Adding a new column for sepal_area
df['sepal_area'] = df['sepal_length'] * df['sepal_width']
print("\n new column added")
print(df.head())
```

```python
def classify_sepal_length(length):
    if length < 5.5:
        return 'Short'
    elif length < 6.5:
        return 'medium'
    else:
        return 'Long'

df['sepal_length_category'] = df['sepal_length'].apply(classify_sepal_length)
print(df.head())
```

```python
pivot_table = pd.pivot_table(df, values='petal_width', index='class', columns='sepal_length_category')
print(pivot_table)
```

```python
encoded_data = pd.get_dummies(df, columns=['class'])
print("\nDataset with class column encoded:")
print(encoded_data.head())
```

```python
# Example: Reshape data using melt
melted_data = pd.melt(df, id_vars=['class'], value_vars=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'])
print("\nMelted dataset:")
print(melted_data.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns 

#scatter plot of petal_length vs sepal_width
plt.figure(figsize=(8,6))
sns.scatterplot(x='sepal_length', y='sepal_width', data=df, hue='class', palette='Set2', s=100)
plt.title('Sepal Length vs Width')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='class', y='petal_length', data=df, palette='Set2')
plt.title('Petal Length by Species')
plt.xlabel('Class')
plt.ylabel('Petal Length')
plt.show()
```
