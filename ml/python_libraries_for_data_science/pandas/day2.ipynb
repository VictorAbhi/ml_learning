```python
data.isnull().sum()
```

```python
data['education'].head(60)
```

```python
# check for missing values in each volumn
missing_data = data.isnull().sum()

# calculate the percentage of missing values
missing_percentage = (missing_data/len(data)) * 100

# combine missing data count and missing percentage
missing_info = pd.DataFrame({
    'Missing Values': missing_data,
    'Missing Percentage': missing_percentage
})

# Filter columns with missing data
missing_info = missing_info[missing_info['Missing Values']>0]

# Format the 'Missing Percentage' column to add a '%' symbol
missing_info['Missing Percentage'] = missing_info['Missing Percentage'].apply(lambda x: f"{x:.2f}%")

# Display the columns with missing data and their percentage
print(missing_info)
```

```python
education_unique_values = data['education'].unique()
previous_year_rating_unique_values = data['previous_year_rating'].unique()

print('Unique values in education:', education_unique_values)
print('Unique values in previous_year_rating_unique_values:', previous_year_rating_unique_values)

```

```python
missing_columns = data.columns[data.isnull().any()]
missing_data_types = data[missing_columns].dtypes

print("\nData types of columns with missing values:")
print(missing_data_types)
```

```python
# Identify numeric and categorical columns
numeric_columns = data.select_dtypes(include=['float64','int64']).columns
categorical_columns = data.select_dtypes(include=['object','category']).columns

# Fill missing values for numeric columns with the mean
data[numeric_columns] = data[numeric_columns].fillna(data[numeric_columns].mean())

for col in categorical_columns:
    data[col] = data[col].fillna(data[col].mode()[0])
df=data.head()
df
```

```python
df['education'].isnull().sum()
```

```python
df.isnull().sum()
```

```python
# convert columns to the specified data types
df['employee_id'] = df['employee_id'].astype('int64')
df['department'] = df['department'].astype('object')
df['region'] = df['region'].astype('object')
df['education'] = df['education'].astype('object')
df['gender'] = df['gender'].astype('object')
df['recruitment_channel'] = df['recruitment_channel'].astype('object')
df['no_of_trainings'] = df['no_of_trainings'].astype('int64')
df['age'] = df['age'].astype('int64')
df['previous_year_rating'] = df['previous_year_rating'].astype('float64')
df['length_of_service'] = df['length_of_service'].astype('int64')
df['KPIs_met_more_than_80'] = df['KPIs_met_more_than_80'].astype('int64')
df['awards_won'] = df['awards_won'].astype('int64')
df['avg_training_score'] = df['avg_training_score'].astype('int64')

print(df.dtypes)
```

```python
df.dtypes
```

```python
duplicates = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicates}")

df = df.drop_duplicates()
print(df.head())
```

```python
def age_group(age):
    if age < 30:
        return 'Youth'
    elif 30 <= age < 50:
        return 'Middle-aged'
    else:
        return 'Senior'

df['age_group'] = df['age'].apply(age_group)
```

```python
df[['age', 'age_group']].head()
```

```python
df['years_at_company'] = df['length_of_service']

def training_score_category(score):
    if score < 50:
        return 'low'
    elif 50 <= score < 75:
        return 'Medium'
    else:
        return 'High'
df['training_score_category'] = df['avg_training_score'].apply(training_score_category)
df[['training_score_category','avg_training_score']]
df.head()
```

```python
def performance_rating(rating):
    if rating >= 4.0:
        return "Excellent"
    elif 3.0 <= rating < 4.0:
        return "Good"
    else:
        return "Needs Improvement"

df['perfomance'] = df['previous_year_rating'].apply(performance_rating)
df['awards_received'] = df['awards_won'].apply(lambda x: True if x >0 else False)
df.head()
```

```python
df['age_and_service'] = df['age'] + df['length_of_service']
df['training_efficiency'] = df['avg_training_score']/df['no_of_trainings']
df['service_to_training_ratio'] = df['length_of_service']/df['no_of_trainings']
df['age_service_interaction'] = df['age'] * df['length_of_service']
df['experience_in_years'] = df['length_of_service']/12
df['education_level'] = df['education'].apply(lambda x: 1 if x in ['Masters','PhD']  else 0)
region_performance = df.groupby('region')['previous_year_rating'].mean().reset_index()
df = df.merge(region_performance, on='region', suffixes=('','_region_avg'))
df['is_young_employee']  = df['age'].apply(lambda x: 1 if x < 30 else 0)

df['is_experienced_employee'] = df['length_of_service'].apply(lambda x: 1  if x > 5 else 0)
df['employee_tenure_in_months'] = df['length_of_service'] * 12
df['age_times_performance'] = df['age'] * df['previous_year_rating']

def service_category(years):
    if years < 2:
        return 'Short Term'
    elif 2 <= years <= 5:
        return 'Medium Term'
    else:
        return 'Long Term'

df['service_category'] = df['length_of_service'].apply(service_category)
df.head()
```
