# Bar Plot

```python
import numpy as np
```

```python
from matplotlib import pyplot as plt
```

```python
student = {"Abhishek Adhikari":100, "Ram Prasad Sharma":69, "Rishi":82}
```

```python
print(student)
```

```python
type(student)
```

```python
names = list(student.keys())
marks = list(student.values())
```

```python
plt.bar(names,marks)
plt.show()
```

## Adding titles and labels to the bar plot

```python
plt.bar(names,marks)
plt.title('Bar Plot')
plt.xlabel('Names')
plt.ylabel('Marks')
plt.show()
```

## Horizontal barplot

```python
plt.barh(names,marks, color='purple')
plt.title('Bar Plot')
plt.xlabel('Names')
plt.ylabel('Marks')
plt.grid(True)
plt.show()
```

## Creating a Basic Scatter Plot

```python
import matplotlib.pyplot as plt
```

```python
x=[10,20,30,40,50,60,70,80,90]
a=[8,1,7,2,0,3,7,3,2]
```

```python
plt.scatter(x,a)
plt.show()
```

```python
plt.scatter(x,a, marker="*", c="red", s=50)
```

## Adding two markers in the same plot

```python
z=[10,20,30,40,50,60,70,80,90]
x=[8,1,7,2,0,3,7,3,2]
y=[7,2,5,6,9,1,4,5,3]
```

```python
plt.scatter(z,x, marker="*", c='g', s=100)
plt.scatter(z,y, marker='.', c='r', s=200)
plt.show()
```

## Adding two subplots

```python
z=[10,20,30,40,50,60,70,80,90]
x=[8,1,7,2,0,3,7,3,2]
y=[7,2,5,6,9,1,4,5,3]

plt.subplot(1,2,1)
plt.scatter(z,x, marker="*", c='g', s=100)

plt.subplot(1,2,2)
plt.scatter(z,y, marker=".", c='r', s=200)
plt.show()
```

# Violin Plot

```python
import matplotlib.pyplot as plt
```

```python
one=[1,2,3,4,5,6,7,8,9]
two=[1,2,3,4,5,4,3,2,1]
three=[6,7,8,9,7,6,5,4]
data=list([one,two,three])
plt.violinplot(data, showmedians=True)
plt.show()
```

# wireframe plot

```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.animation as animation

x=np.linspace(-5,5,100)
y=np.linspace(-5,5,100)
X,Y=np.meshgrid(x,y)
Z=np.sin(np.sqrt(X**2+Y**2))

fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')
wireframe = ax.plot_wireframe(X,Y,Z)

ax.set_title('Self-Rotating 3D Wireframe Plot')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')

def update(frame):
    ax.clear()
    new_Z = np.sin(np.sqrt((X+frame/50)**2+(Y+frame/50)**2))
    wireframe = ax.plot_wireframe(X,Y,new_Z,color='b')
    ax.set_title('Self-Rotating 3D Wireframe Plot')
    ax.set_xlabel('X-axis')
    ax.set_ylabel('Y-axis')
    ax.set_zlabel('Z-axis')
    ax.view_init(elev=30, azim=frame)
    return wireframe,

ani = animation.FuncAnimation(fig,update,frames=np.arange(0,360,2),interval=50, blit=True)
plt.show()
```
