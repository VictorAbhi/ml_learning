# Line Plot Basics

```python
import numpy as np
from matplotlib import pyplot as plt
# import matplotlib.pyplot as plt
```

```python
x = np.arange(1,11)
```

```python
print(x)
```

```python
y = 2*x
```

```python
print(y)
```

```python
plt.plot(x,y)
# plt.show()
```

### Adding Titles and lables

```python
plt.plot(x,y)
plt.title("Line Plot of x and y variable")
plt.xlabel('x-label')
plt.ylabel('y-label')
plt.show()
```

### Changing Line Aesthetics

```python
plt.plot(x,y,color='g', linestyle=":", linewidth=3)
plt.title("Line Plot of x and y variable")
plt.xlabel('x-label')
plt.ylabel('y-label')
plt.show()
```

## Adding two lines in the same plot

```python
y1=2*x
y2=3*x
```

```python
plt.plot(x,y1,color='g',linewidth=3)
plt.plot(x,y2,color='r', linestyle=":", linewidth=5)
plt.grid(True)
```

## Adding subplots horizontally

```python
plt.subplot(1,6,1)
plt.plot(x,y1,color='g',linewidth=3)

plt.subplot(1,6,2)
plt.plot(x,y1,color="r", linewidth=3)

plt.subplot(1,6,3)
plt.plot(x,y1,color='g',linewidth=3)

plt.subplot(1,6,4)
plt.plot(x,y1,color="r", linewidth=3)

plt.subplot(1,6,5)
plt.plot(x,y1,color='g',linewidth=3)

plt.subplot(1,6,6)
plt.plot(x,y1,color="r", linewidth=3)
plt.show()
```

## Adding subplots vertically

```python
plt.subplot(2,1,1)
plt.plot(x,y1,color='r', linewidth=3)

plt.subplot(2,1,2)
plt.plot(x,y1,color='r', linewidth=3)
plt.show()
```

```python
import pandas as pd
```

```python
df = pd.read_csv('../pandas/data_sets/number.csv')
```

```python
def function(number):
    if number['x']> 1 or number['y'] > 1 or number['z'] > 1:
        return 3*number['x'] + 2 * np.exp(number['y']) + 5 * number['z']
    else:
        return 0

df['p'] = df.apply(function, axis=1)
```

```python
df.head()
```

```python
df['p'] = df.apply(lambda number:3*number['x'] + 2 * np.exp(number['y']) + 5 * number['z'] if  number['x']> 1 or number['y'] > 1 or number['z'] > 1 else 0, axis=1)
```

```python
df.head()
```

```python
# Plot p vs x
plt.figure(figsize=(10,6))
plt.plot(df['x'], df['p'], marker='o', linestyle='-', color='b')
plt.title('Line plot of p vs x')
plt.xlabel('p')
plt.ylabel('x')
plt.grid(True)
plt.show()

# Plot p vs y
plt.figure(figsize=(10,6))
plt.plot(df['y'], df['p'], marker='*', linestyle='-', color='b')
plt.title('Line plot of p vs y')
plt.xlabel('p')
plt.ylabel('y')
plt.grid(True)
plt.show()

# Plot p vs z
plt.figure(figsize=(10,6))
plt.plot(df['p'], df['z'], marker='o', linestyle='-', color='b')
plt.title('Line plot of p vs z')
plt.xlabel('p')
plt.ylabel('z')
plt.grid(True)
plt.show()


```

# Making histogram

```python
data = [1,3,3,3,3,3,9,9,4,4,8,8,6,7]
```

```python
plt.hist(data)
plt.show()
```

### Adding with the bins

```python
plt.figure(figsize=(9,2))
plt.hist(data, color='b', bins=3)
plt.show()
```

```python
iris_df = pd.read_csv('../pandas/data_sets/iris.csv')
```

```python
iris_df.head()
```

```python
iris_df = iris_df.rename(columns={'class': 'species'})
```

```python
plt.figure(figsize=(5,2))
plt.subplot(1,4,1)
plt.hist(iris_df['sepal_length'],bins=30, color='g')

plt.subplot(1,4,2)
plt.hist(iris_df['petal_length'],bins=30, color='r')

plt.subplot(1,4,3)
plt.hist(iris_df['sepal_width'],bins=30, color='b')

plt.subplot(1,4,4)
plt.hist(iris_df['sepal_length'],bins=30, color='y')

plt.show()
```
