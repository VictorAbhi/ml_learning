```python
import sqlite3

# Load CSV file into a Pandas DataFrame
csv_file = "athlete_events.csv"  # Replace with your actual CSV file name
df1 = pd.read_csv(csv_file)

# Connect to SQLite database (or create it if it doesn't exist)
conn = sqlite3.connect("sqlite.db")  # Creates an SQLite database file
cursor = conn.cursor()

# Create the transactions table
cursor.execute("""
CREATE TABLE IF NOT EXISTS Athlete (
    ID INTEGER PRIMARY KEY,
    Name TEXT,
    Sex TEXT,
    Age INTEGER,
    Height REAL,
    Weight REAL,
    Team TEXT,
    NOC TEXT,
    Games TEXT,
    Year INTEGER,
    Season TEXT,
    City TEXT,
    Sport TEXT,
    Event TEXT,
    Medal TEXT
);
""")


# Insert the data into SQLite
df1.to_sql("athlete_events", conn, if_exists="replace", index=False)

# Commit and close the connection
conn.commit()
conn.close()

print("CSV file has been successfully imported into the SQLite database!")
```

```python
%%sql
SELECT ae.Games, COUNT(nr.region) AS total_no_of_countries
FROM athelete_events AS ae
INNER JOIN noc_regions AS nr ON ae.NOC = nr.NOC
GROUP BY ae.Games
LIMIT 5;
```

```python
%%sql
SELECT ae.games, COUNT(DISTINCT nr.region) AS total_countries
FROM athelete_events ae
JOIN noc_regions nr ON nr.noc = ae.noc
GROUP BY ae.games
ORDER BY ae.games;
```

```python
%%sql
WITH game_region_counts AS (
    SELECT ae.games, COUNT(DISTINCT nr.region) AS total_countries
    FROM athelete_events ae
    JOIN noc_regions nr ON nr.noc = ae.noc
    GROUP BY ae.games
)

SELECT games, total_countries AS lowest_countries
FROM game_region_counts
WHERE total_countries = (SELECT MIN(total_countries) FROM game_region_counts)
UNION
SELECT games, total_countries AS highest_countries
FROM game_region_counts
WHERE total_countries = (SELECT MAX(total_countries) FROM game_region_counts);
```

```python
%%sql
with country_participation as (
    select nr.region as country, count(distinct games) as game_participated
    from athelete_events ae
    join noc_regions nr on nr.NOC=ae.NOC
    group by nr.region
)

-- countries participated in every games
select country
from country_participation
where game_participated = (select count(distinct games) from athelete_events);
```

```python
%%sql
-- no of sports part in every summer oplympics
with games_played as (
    select Sport, count(distinct Season) as season_count, count(distinct games) as total_games_played
    from athelete_events
    where Season='Summer'
    group by Sport
)

select sport, total_games_played
from games_played
where total_games_played = (select count(distinct games) from athelete_events
                                    where season='Summer' )
```

```python
%%sql
-- games played only once in the olympics
with games_played_once as (
    select sport, count(distinct games) as games_played
    from athelete_events
    group by sport
    having games_played = 1
)
select * from games_played_once;
```

```python
%%sql
-- total no of sports played in each olympics
select games, count(distinct sport) as no_of_sports 
from athelete_events
group by games
order by no_of_sports desc;
```

```python
%%sql
-- oldest athlete to win the gold medal
select name, sex, age, team, sport
from athelete_events
where Medal='Gold'
and age=(select max(age) from athelete_events where medal='Gold');
```

```python
%%sql
-- Ratio of males and females participated in Olympic Games
WITH male_female AS (
    SELECT
        SUM(CASE WHEN sex = 'M' THEN 1 ELSE 0 END) AS male_count,
        SUM(CASE WHEN sex = 'F' THEN 1 ELSE 0 END) AS female_count
    FROM athelete_events
)
SELECT 
    ROUND(CAST(male_count AS FLOAT) / NULLIF(female_count, 0), 2) AS male_to_female_ratio
FROM male_female;
```

```python
%%sql
-- top 5 atheletes winning the most gold medals
SELECT name, sport, COUNT(*) AS gold_medals
FROM athelete_events
WHERE Medal = 'Gold'
GROUP BY name, sport
ORDER BY gold_medals DESC
LIMIT 5;
```

```python
%%sql
-- top 5 atheletes winning the most  medals
SELECT name, sport, COUNT(Medal) AS medals
FROM athelete_events
WHERE Medal is not null
GROUP BY name, sport
ORDER BY medals DESC
LIMIT 5;
```

```python
%%sql
-- top 5 successful countries in olympics
select nc.region, COUNT(ae.Medal) as medals
from noc_regions as nc
join athelete_events as ae on nc.NOC=ae.NOC
WHERE ae.Medal is not null
GROUP BY nc.region
order by medals desc limit 5;
```

```python
%%sql
-- Total Gold, Silver, and Bronze Medals by each country
SELECT 
    nc.region,
    COUNT(CASE WHEN ae.Medal = 'Gold' THEN 1 END) AS gold_medals,
    COUNT(CASE WHEN ae.Medal = 'Silver' THEN 1 END) AS silver_medals,
    COUNT(CASE WHEN ae.Medal = 'Bronze' THEN 1 END) AS bronze_medals
FROM noc_regions AS nc
JOIN athelete_events AS ae ON nc.NOC = ae.NOC
WHERE ae.Medal IS NOT NULL
GROUP BY nc.region
ORDER BY gold_medals DESC, silver_medals DESC, bronze_medals DESC;
```

```python
%%sql
-- Total Gold, Silver, and Bronze Medals by each country
SELECT 
    ae.games,nc.region as country,
    COUNT(CASE WHEN ae.Medal = 'Gold' THEN 1 END) AS gold_medals,
    COUNT(CASE WHEN ae.Medal = 'Silver' THEN 1 END) AS silver_medals,
    COUNT(CASE WHEN ae.Medal = 'Bronze' THEN 1 END) AS bronze_medals
FROM noc_regions AS nc
JOIN athelete_events AS ae ON nc.NOC = ae.NOC
WHERE ae.Medal IS NOT NULL
GROUP BY ae.games, nc.region
ORDER BY ae.games, nc.region limit 10;
```
